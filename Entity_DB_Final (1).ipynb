```python
# Define your lists (same as in your code)

# Take input from the user
user_input = input("Enter a sentence: ")

# Function to find matches in a list using fuzzy matching
def find_fuzzy_matches(input_text, search_list, threshold=96):
    matches = []
    for item in search_list:
        # Calculate the similarity score between the input text and the item in the list
        similarity_score = fuzz.partial_ratio(input_text.lower(), item.lower())
        if similarity_score >= threshold:
            matches.append(item)
    return matches

# Find matches in player names using fuzzy matching
matched_players = find_fuzzy_matches(user_input, player_names)
matched_teams = find_fuzzy_matches(user_input, team_names)
matched_columns = find_fuzzy_matches(user_input, columns)

```

```python
player_parameter=''
for i in matched_players:
    player_parameter=player_parameter+i+'|'
    
player_parameter=player_parameter[0:len(player_parameter)-1]


team_parameter=''
for i in matched_teams:
    team_parameter=team_parameter+i+'|'
    
team_parameter=team_parameter[0:len(team_parameter)-1]


attribute_parameter=''
for i in matched_columns:
    attribute_parameter=attribute_parameter+i+'|'
    
attribute_parameter=attribute_parameter[0:len(attribute_parameter)-1]

```

```python
# Define the server and database information
server = 'NSPLUS-LP26\\SQLEXPRESS'
database = 'BPL'

# Create a connection string
connection_string = f'DRIVER={{SQL Server}};SERVER={server};DATABASE={database}'

# Establish a connection
conn = pyodbc.connect(connection_string)
cursor = conn.cursor()
```

```python
import pandas as pd

# Create an empty list to store the dataframes
dataframes = []

while True:
    # Check if the current statement is a query
    if cursor.description is not None:
        # Fetch all rows from the current result set
        rows = cursor.fetchall()
        result_2=[]
        for i in rows:
            result_2.append(tuple(i))
            com = [item[0] for item in cursor.description]

            df = pd.DataFrame(result_2, columns=com)

        
        
        
        # Append the dataframe to the list
        dataframes.append(df)
    
    # Move to the next result set, if available
    if not cursor.nextset():
        break

# Close the cursor and connection
cursor.close()
conn.close()

# Access the individual dataframes from the list
df1 = dataframes[0]
df2 = dataframes[1]
df3 = dataframes[2]
df4 = dataframes[3]
df5 = dataframes[4]
```
